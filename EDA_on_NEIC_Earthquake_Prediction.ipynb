```python
# Converting Date to DateTime
data['Date'] = pd.to_datetime(data['Date'], utc=True)
#data['Date'] = data['Date'].dt.strftime('%Y-%m-%d')

# Making Float for data['Depth Seismic Stations'] :- 
data['Depth Seismic Stations'] = pd.to_numeric(data['Depth Seismic Stations'])

# Making Float for data['Depth Seismic Stations'] :- 
data['Root Mean Square'] = pd.to_numeric(data['Depth Seismic Stations'])
```

```python
# Filling the missing values of Depth Error feature with the mean
data['Depth Error'].fillna(value=data['Depth Error'].mean(), inplace=True)

# Filling the missing values of Depth Error feature with the mean
data['Depth Seismic Stations'].fillna(value=data['Depth Seismic Stations'].mean(), inplace=True)

# Filling the missing values of Depth Error feature with the mean
data['Root Mean Square'].fillna(value=data['Root Mean Square'].mean(), inplace=True)

# Filling the missing values of Depth Error feature with the mean 
data['Magnitude Type'].fillna(value=data['Magnitude Type'].mode()[0], inplace=True)
```
